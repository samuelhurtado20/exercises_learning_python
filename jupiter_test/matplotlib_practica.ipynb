# Punto 0:

```python
# importar pandas
import pandas as pd
```

```python
# datos de RRSS 2020
fbk = ['Facebook', 2449, True, 2006]
twt = ['Twitter', 339, False, 2006]
ig = ['Instagram', 1000, True, 2010]
yt = ['YouTube', 2000, False, 2005]
lkn = ['LinkedIn', 663, False, 2003]
wsp = ['WhatsApp', 1600, True, 2009]

# lista con datos
lista_rrss = [fbk, twt, ig, yt, lkn, wsp]

# crear dataframe a partir de listas
df =pd.DataFrame(lista_rrss,
            columns = ['Nombre', 'Cantidad', 'ES_FBK', 'A침o'])
df
```

```python
# matplotlib basics 101
```

```python
# importar matplotlib
import matplotlib.pyplot as plt
```

```python
# ejemplo super mega b치sico
x=[1,2,3,4,5]
y=[x ** 3 for x in x]
print(y)
plt.plot(x,y)
plt.show()
```

# Punto 1:

```python
# gr치fico de l칤neas utilizando matplotlib
plt.plot(df['Nombre'], df['Cantidad'])
plt.show()
```

```python
# gr치fico de puntos utilizando matplotlib
plt.scatter(df['Nombre'], df['Cantidad'])
plt.show()
```

```python
# EXTRA: c칩mo hacer lo mismo con m칠todos de un dataframe de pandas
df.plot(kind='scatter', x='Nombre', y='Cantidad')
```

# Punto 2:

```python
# hacer gr치ficos de barras
```

```python
# gr치fico de puntos utilizando matplotlib
plt.bar(df['Nombre'], df['Cantidad'])
plt.show()
```

```python
# gr치fico de barras ordenado
df_sorted = df.sort_values('Cantidad', ascending=True)
plt.bar(df_sorted['Nombre'], df_sorted['Cantidad'])
plt.show()
```

```python
# gr치fico de barras ordenado
df_sorted = df.sort_values('Cantidad', ascending=False)
plt.bar(df_sorted['Nombre'], df_sorted['Cantidad'])
plt.show()
```

```python
# gr치fico de barras ordenado y con color


# b: blue.
# g: green.
# r: red.
# c: cyan.
# m: magenta.
# y: yellow.
# k: black.
# w: white.
df_sorted = df.sort_values('Cantidad', ascending=False)
plt.bar(df_sorted['Nombre'], df_sorted['Cantidad'], color=['#3b5998', '#FF0000', '#25d366', '#8a3ab9', '#0e76a8', '#00acee'])
plt.show()
```
